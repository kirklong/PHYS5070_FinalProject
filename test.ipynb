```python
AU = 1.5e11; Msun = 2e30; G = 6.67e-11
v = np.sqrt(G*Msun/AU)
Earth = body([AU,0.,0.],[0.,v,0.],6e24)
Sun = body([0.,0.,0.,],[0.,0.,0.,],Msun)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig,ax = plt.subplots(1,figsize=(10,10),facecolor="white")
ax.set_xlim((-1.5,1.5))
ax.set_ylim((-1.5,1.5))
ax.set_xlabel("x [AU]")
ax.set_ylabel("y [AU]")

XEarth = []; YEarth = []
XSun = []; YSun = []
s = system([Sun,Earth],Î”t=yearSec/1e3,massiveInd=0,closeInds=[1]) #initialize
while s.T < 1:
    s.update()
    x,y,z,vx,vy,vz = s.getCoords()
    XSun.append(x[0]); YSun.append(y[0])
    XEarth.append(x[1]); YEarth.append(y[1])

ax.plot(np.array(XEarth)/AU,np.array(YEarth)/AU,c="dodgerblue",marker=".",ms=1)
ax.plot(np.array(XSun)/AU,np.array(YSun)/AU,c="gold",marker="o")
ax.set_facecolor("k")
```
